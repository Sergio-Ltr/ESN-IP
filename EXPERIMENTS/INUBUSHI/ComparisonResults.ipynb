```python
import pandas as pd
```

```python
result_df = pd.read_csv(filepath_or_buffer='./INUBUSHI-DIST-COMP_NU_1_TAU_1.csv')
```

```python
result_df.reset_index(drop=True)
```

```python
# Split DataFrame into even and odd rows
df_mean = result_df[result_df.index % 2 == 0].reset_index(drop=True)
df_std = result_df[result_df.index % 2 != 0].reset_index(drop=True)
```

```python
df_mean = df_mean[df_mean['Model Index'].isin([1, 2, 4])]
df_std = df_std[df_std['Model Index'].isin([1, 2, 4])]
```

```python
model_names = ["Vanilla", "Gaussian", "Hard Bimodal", "Soft Trimodal", "Edgy trimodal" ]
```

```python
mask_names = [ "Gaussian", "Bimodal", "Trimodal" ]
```

## MSE PLOT

```python
import seaborn as sns
```

```python
custom_palette = sns.color_palette("Set2")
```

```python
custom_palette
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Example DataFrame structure:
# df = pd.DataFrame({'N': ..., 'NRMSE_mean': ..., 'NRMSE_std': ..., 'model_index': ...})

# Set the style if needed
sns.set(style="whitegrid")

# Adjust figure size if needed
plt.figure(figsize=(10, 6))


indices_to_plot = [1,2,4]
colors = custom_palette[2:5]

# Use Seaborn's lineplot() with error bars to plot the mean values and standard deviation
sns.lineplot(x='N', y='NRMSE', hue=df_mean.columns[3], data=df_mean, marker='o', palette=colors)


# Use matplotlib's errorbar to plot the standard deviation with the same colors
for i, color in enumerate(colors):
    plt.errorbar(x=df_std[df_std['Model Index'] == indices_to_plot[i]]['N'], 
                 y=df_mean[df_mean['Model Index'] == indices_to_plot[i]]['NRMSE'], 
                 yerr=df_std[df_std['Model Index'] == indices_to_plot[i]]['NRMSE'], 
                 fmt='none', capsize=5, c=color, alpha=0.7)

plt.errorbar(x=df_std['N'], y=df_mean['NRMSE'], yerr=df_std['NRMSE'], fmt='none', capsize=5, c='gray', alpha=0.3)

plt.title('Error vs. Complexity')
plt.xlabel('Number of Units (N)')
plt.ylabel('NRMSE (Performance)')
plt.legend(title='', labels=mask_names)

plt.show()
```

## DELTA PHI PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Example DataFrame structure:
# df = pd.DataFrame({'N': ..., 'DeltaPhi_mean': ..., 'DeltaPhi_std': ..., 'model_index': ...})

# Set the style if needed
sns.set(style="whitegrid")

# Adjust figure size if needed
plt.figure(figsize=(10, 6))


indices_to_plot = [1,2,4]
colors = custom_palette[2:5]

# Use Seaborn's lineplot() with error bars to plot the mean values and standard deviation
sns.lineplot(x='N', y='DeltaPhi', hue=df_mean.columns[3], data=df_mean, marker='o', palette=colors)


# Use matplotlib's errorbar to plot the standard deviation with the same colors
for i, color in enumerate(colors):
    plt.errorbar(x=df_std[df_std['Model Index'] == indices_to_plot[i]]['N'], 
                 y=df_mean[df_mean['Model Index'] == indices_to_plot[i]]['DeltaPhi'], 
                 yerr=df_std[df_std['Model Index'] == indices_to_plot[i]]['DeltaPhi'], 
                 fmt='none', capsize=5, c=color, alpha=0.7)

plt.errorbar(x=df_std['N'], y=df_mean['DeltaPhi'], yerr=df_std['DeltaPhi'], fmt='none', capsize=5, c='gray', alpha=0.3)

plt.title('Non Linearity vs. Complexity')
plt.xlabel('Number of Units (N)')
plt.ylabel('DeltaPhi (Non Linearity)')
plt.legend(title='Model Index', labels=mask_names)

plt.show()
```
