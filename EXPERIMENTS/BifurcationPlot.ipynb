```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def bifurcation_diagram(f, x0, r_range, num_iterations, num_transient, num_points, ax=None, **kwargs):
    """
    Plot a bifurcation diagram of a dynamical system.

    Parameters:
        f: function
            The dynamical system function. It should take two arguments: x and r.
        x0: float
            Initial condition for the system.
        r_range: tuple
            Range of parameter values (r) to explore.
        num_iterations: int
            Number of iterations to perform for each parameter value.
        num_transient: int
            Number of transient iterations to discard.
        num_points: int
            Number of points to sample within the parameter range.
        ax: matplotlib axis, optional
            The axis to plot on. If None, a new figure and axis will be created.
        **kwargs:
            Additional keyword arguments to be passed to sns.scatterplot.

    Returns:
        ax: matplotlib axis
            The axis containing the bifurcation diagram.
    """
    # Initialize arrays to store parameter and state values
    r_values = np.linspace(r_range[0], r_range[1], num_points)
    x_values = np.zeros((num_iterations, num_points))

    # Iterate over parameter values
    for i, r in enumerate(r_values):
        x = x0
        # Discard transient iterations
        for _ in range(num_transient):
            x = f(x, r)
        # Store state values
        for j in range(num_iterations):
            x = f(x, r)
            x_values[j, i] = x

    # Plot bifurcation diagram
    if ax is None:
        fig, ax = plt.subplots()
    sns.scatterplot(x=r_values, y=x_values[-1], ax=ax, **kwargs)
    ax.set_xlabel('Parameter (r)')
    ax.set_ylabel('State (x)')
    ax.set_title('Bifurcation Diagram')
    return ax

# Example dynamical system: Logistic map
def logistic_map(x, r):
    return r * x * (1 - x)

# Parameters
x0 = 0.5  # Initial condition
r_range = (2.5, 4.0)  # Range of parameter values (r)
num_iterations = 1000  # Number of iterations for each parameter value
num_transient = 100  # Number of transient iterations to discard
num_points = 1000  # Number of points to sample within the parameter range

# Plot bifurcation diagram
sns.set(style='whitegrid')  # Set seaborn style
bifurcation_diagram(logistic_map, x0, r_range, num_iterations, num_transient, num_points)
plt.show()
```
