```python

def results_timeseries_plotter(attribute = "MSE", filepath = "../BENCHMARKS/RESULTS/4.3_RESULTS_Inubushi_ni=0.5.csv", title = "Inubushi Function (ni = 0.5)", logscale = True):
    
    df = pd.read_csv(filepath)
    unit_column = np.array([np.ones(35)*100, np.ones(35)*150, np.ones(35)*200, np.ones(35)*250]).flatten()
    df.insert(loc=2, column='UNITS', value=unit_column)
    dataframe = pd.DataFrame(df.groupby(["UNITS", "TYPE"]).mean()[attribute])

    # Reset index to make 'UNITS' and 'TYPE' columns
    df_reset = dataframe.reset_index()

    # Create a multi-index
    df_multiindex = df_reset.set_index(['UNITS', 'TYPE'])

    # Unstack 'TYPE' to columns
    df_unstacked = df_multiindex.unstack()

    # Plot the data
    ax = df_unstacked.plot(kind='line', marker='o', logy=logscale)

    # Set labels and title
    plt.xlabel('UNITS')
    plt.ylabel(f'{attribute} {"(logscale)" if logscale else ""}')
    plt.title(title)

    # Customize legend labels
    legend_labels = [
        'Vanilla',
        'Gauss 1',
        'Gauss 2',
        'Bimodal 3',
        'Trim 1',
        'Trim 2',
        'TRIM 3'
    ]

    ax.legend(labels= [legend_labels[i] for i in range(df_unstacked.columns.shape[0])])

    # Show the plot
    plt.show()
```

```python
attribute = "MSE"
unit_column = np.array([np.ones(6)*100, np.ones(6)*150, np.ones(6)*200, np.ones(6)*250]).flatten()
df.insert(loc=2, column='UNITS', value=unit_column)
dataframe = pd.DataFrame(df.groupby(["UNITS", "TYPE"]).mean()[attribute])

# Reset index to make 'UNITS' and 'TYPE' columns
df_reset = dataframe.reset_index()

# Create a multi-index
df_multiindex = df_reset.set_index(['UNITS', 'TYPE'])

# Unstack 'TYPE' to columns
df_unstacked = df_multiindex.unstack()

# Plot the data
ax = df_unstacked.plot(kind='line', marker='o', logy=True)

# Set labels and title
plt.xlabel('UNITS')
plt.ylabel(f'{attribute} (logscale) ')
plt.title("MSE on Dambre non linear function")

# Customize legend labels
legend_labels = [
    'Vanilla',
    'Gauss 1',
    'Gauss 2',
    'Bimodal 3',
    'Trim 1',
    'Trim 2'
]

ax.legend(labels= [legend_labels[i] for i in range(df_unstacked.columns.shape[0])])

# Show the plot
plt.show()
```
