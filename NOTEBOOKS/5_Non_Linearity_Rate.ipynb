```python
from ESN import Reservoir

reservoir = Reservoir(1,100, 0.1, 0.9, bias = True)
```

```python
f_range = np.linspace(0.01, 0.501, num=101)*500
```

```python
f_range = np.linspace(0.01, 0.5, num=101)*500
```

```python
f_range = f_range[0:-1]
```

```python
f_range
```

```python
starttime = 0.0
endtime = 2.0
steps = 1000
t = np.linspace(starttime, endtime, num=steps)
```

```python
f_range.shape
```

```python
f_range
```

```python
de_fi_acc = 0

for f in f_range:
    samples = 1000

    x = np.arange(0, samples)
    y = np.sin(2 * np.pi * int(f) * t)

    y_Res = reservoir.predict(y).numpy()
    reservoir.reset_initial_state()

    dc = np.mean(y_Res, axis=0 )

    y_Res = y_Res - dc

    fhat = np.fft.fft(np.mean(y_Res, axis = 1))
    N = len(fhat)
    halvedfhat = fhat[0:int(N/2)]
    powspec = abs(halvedfhat)**2

    print(max(powspec), powspec[int(f)*2])

    fs = (steps)/(endtime - starttime)

    freq = np.linspace(0,int(fs/2),int(N/2))

    plt.plot(freq, powspec)
    e_tot = np.sum(powspec)

    de_fi = 1 - powspec[int(f)*2]/e_tot
    print(int(f), de_fi)
    de_fi_acc += de_fi

print(de_fi_acc/len(f_range))
```
