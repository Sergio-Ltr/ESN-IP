## Setup the environment

Idea is that as the number of hidden units of a Reservoir increases, the distribution of their output, before the application of the non-linearity, is a Gaussian distribution. 

If such hypothesis would be proven, it would mean that the "classic" IP algorithm (targeting normal distributions), would converge faster. 
Hence it would be useful to discover which reservoir hyperparameter configuration would lead to a "more Gaussian" distribution and hence to faster convergence.

```python
from IPReservoir import IPReservoir
from IntrinsicPlasticity import IPMask
```

```python
import itertools  

sparsity_range = [0.1, 0.25, 0.5, 0.75, 0.9]
ro_range = [0.25, 0.5, 0.75, 1, 1.25]

combinations = list(itertools.product(sparsity_range, ro_range))
len(combinations)
```

```python
for c in combinations: 
    s, r = c
    print(f"Units: 50, Sparsity: {s}, Spectral Radius of recurrent weights matrix: {r}")
    print("MG DATA DISTRIBUTION")
    ip_reservoir = IPReservoir(sparsity=s, ro_rescale=r, N = 50, mask=IPMask.normalMask(50), bias = True)
    ip_reservoir.pre_train(X_VAL_MG, eta=0.0000, epochs=1,  learning_rule="online", verbose=False) #16
    ip_reservoir.predict(X_VAL_MG, save_gradients=False, save_states=True)
    ip_reservoir.plot_global_neural_activity()
    print("NARMA DATA DATA DISTRIBUTION")

    ip_reservoir.sample_targets(X_VAL_NARMA.shape[0], overwrite=True)
    ip_reservoir.predict(X_VAL_NARMA, save_gradients=False, save_states=True)
    ip_reservoir.plot_global_neural_activity()
```
